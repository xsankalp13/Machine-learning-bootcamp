```python
#increasing the tax 5% by 2 times

df['Tax 5%'] = df['Tax 5%'].apply(lambda x: x*2)
df.head()
```

```python
#changing the column name Tax 5% to Tax 10%

df.rename(columns={'Tax 5%' : 'Tax 10%'}, inplace=True)
df.head()
```

```python
#updating the Total value

df['Total'] = df['Unit price']* df['Quantity'] + df['Tax 10%']
df.head()
```

```python
grouped_mean = df.groupby(['Product line','Gender'])['Total'].mean()
print(grouped_mean)
```
