```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#basic seaborn ploting
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter plot of total bill and tip')
plt.show()
```

```python
##line plot
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title('Line plot of total bill and size')
plt.show()
```

```python
##bar plot
sns.barplot(x='day',y='total_bill', data=tips)
plt.title('Bar plot of total bill and day')
plt.show()
```

```python
##Box plot
sns.boxplot(x='day',y='total_bill', data=tips)
plt.title('Box plot of total bill and day')
plt.show()
```

```python
##Violin plot
sns.violinplot(x='day',y='total_bill', data=tips)
plt.title('Violin plot of total bill and day')
plt.show()
```

```python
##histogram
sns.histplot(tips['total_bill'], bins=10, kde=True)
plt.title('Histogram of total bill')
plt.show()
```

```python
#KDE plot
sns.kdeplot(tips['total_bill'], fill=True)
plt.title('KDE plot of total bill')
plt.show()
```

```python
#pair plot

sns.pairplot(tips)
plt.title('Pair plot of tips')
plt.show()
```

```python
##heatmap

corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
